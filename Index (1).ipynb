# CUSTOMER SATISFACTION IN AN AIRLINE

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('Invistico_Airline.csv')
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df['satisfaction']=l.fit_transform(df['satisfaction'])
df.head(5)
```

```python
df.isnull().any()
```

```python
x=df[['Departure/Arrival time convenient','Food and drink','Inflight wifi service','Online support','Ease of Online booking','On-board service','Baggage handling','Checkin service','Cleanliness','Online boarding','Departure Delay in Minutes']]
x.head(5)
```

```python
from sklearn.model_selection import train_test_split
x1,x2,y1,y2=train_test_split(x,df['satisfaction'],test_size=0.2)
```

```python
from sklearn import tree
reg=tree.DecisionTreeClassifier()
reg.fit(x1,y1)
```

```python
reg.predict(x2)
```

```python
reg.score(x2,y2)
```
